```python
#performing label encoding

from sklearn.preprocessing import LabelEncoder
```

```python
la=LabelEncoder()
data['tartar']=la.fit_transform(data['tartar'])
data['dental caries']=la.fit_transform(data['dental caries'])
data['gender']=la.fit_transform(data['gender'])
```

```python
#dependendt and independent variables
x=data.iloc[:,:-1]
y=data.loc[:,'smoking']
```

```python
y.ndim
```

```python
#feature selection using important features
'''Feature importance is a technique
that calculate a score for all the
input features for a given model. So
out of 24 features we will select the
top 15 features based on the score.'''

from sklearn.ensemble import ExtraTreesClassifier
model=ExtraTreesClassifier()
model.fit(x,y)
df1=pd.Series(model.feature_importances_,index=x.columns)
plt.figure(figsize=(8,8))
df1.nlargest(24).plot(kind='barh')
plt.show()
```

```python
#splitting the data into train and test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42)
```

```python
#standarlizing the data
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
#model building
#logistic regression
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
#decision tree
from sklearn.tree import DecisionTreeClassifier
```

```python
dt=DecisionTreeClassifier()
```

```python
dt.fit(x_train,y_train)
```

```python
y_pred=dt.predict(x_test)
```

```python
print(classification_report(y_test,y_pred))
```

```python
#Bagging
from sklearn.ensemble import BaggingClassifier
bc=BaggingClassifier(base_estimator=DecisionTreeClassifier(),n_estimators=1000)
bc.fit(x_train,y_train).score(x_test,y_test)
y_pred=bc.predict(x_test)
```

```python
print(classification_report(y_test,y_pred))
```

```python
from sklearn.ensemble import ExtraTreesClassifier
et=ExtraTreesClassifier(n_estimators=1000,random_state=42)
et.fit(x_train,y_train)
y_pred=et.predict(x_test)
print(classification_report(y_test,y_pred))
```

```python
#random forest
from sklearn.ensemble import RandomForestClassifier
```

```python
rf=RandomForestClassifier()
rf.fit(x_train,y_train)
y_pred=rf.predict(x_test)
```

```python
print(classification_report(y_test,y_pred))
```
